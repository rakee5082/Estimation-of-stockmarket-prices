```python
from subprocess import Popen, PIPE
import sys


# proc = Popen([sys.argv[1]], stdin=PIPE, stdout=PIPE)
proc = Popen(['python', 'sample_submit.py'], stdin=PIPE, stdout=PIPE)

n = 10
proc.stdin.write(('%d\n' % n).encode('utf-8'))

currs = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']


fs = {}
DATA_FOLDER = 'data'
for c in currs:
    fs['{}_ticker'.format(c)] = open(DATA_FOLDER + '/{}_ticker.csv'.format(c))
    fs['{}_trades'.format(c)] = open(DATA_FOLDER + '/{}_trades.csv'.format(c))
    fs['{}_book'.format(c)] = open(DATA_FOLDER + '/{}_book.csv'.format(c))

for k in range(n):
    for c in currs:
        l = fs['{}_ticker'.format(c)].readline()
        proc.stdin.write((l).encode('utf-8'))
        proc.stdin.flush()
        
        l = fs['{}_trades'.format(c)].readline()
        proc.stdin.write((l).encode('utf-8'))
        proc.stdin.flush()

        l = fs['{}_book'.format(c)].readline()
        proc.stdin.write((l).encode('utf-8'))
        proc.stdin.flush()

    results = []
    for j in range(len(currs)):
        res = []
        i = ''
        while i != '\n'.encode('utf-8'):
            i = proc.stdout.read(1)
            res.append(i)
        results.append(b''.join(res).decode('utf-8').strip())
    print(results)
```
